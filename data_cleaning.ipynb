```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
#Steps to make the data messy.
#The original dataset is pretty clean, to practice the data cleaning steps, we can make the data a bit messy using these steps

hotel_bookings_clean = pd.read_csv('hotel_bookings.csv')

hotel_bookings_messy = hotel_bookings_clean.copy()

#Change some of the int dtypes to floats and some to strings
hotel_bookings_messy['lead_time'] = hotel_bookings_messy['lead_time'].astype('str')

#Combine arrival date and month and drop separate columns
hotel_bookings_messy['arrival_date'] = hotel_bookings_messy['arrival_date_month'] + "-" + hotel_bookings_messy['arrival_date_year'].astype('str')
hotel_bookings_messy.drop(columns = ['arrival_date_month', 'arrival_date_year'], inplace=True)

column_to_move = hotel_bookings_messy.pop('arrival_date')
hotel_bookings_messy.insert(3, 'arrival_date', column_to_move)

#Add random special characters
indexes = hotel_bookings_messy['hotel'].sample(10000).index.tolist()
character_to_add = '^'
hotel_bookings_messy.loc[indexes, 'hotel'] = character_to_add + hotel_bookings_messy['hotel']


indexes = hotel_bookings_messy['hotel'].sample(10000).index.tolist()
character_to_add = '**'
hotel_bookings_messy.loc[indexes, 'hotel'] = hotel_bookings_messy['hotel'] + character_to_add 

indexes = hotel_bookings_messy['hotel'].sample(10000).index.tolist()
character_to_add = '\n'
hotel_bookings_messy.loc[indexes, 'hotel'] = hotel_bookings_messy['hotel'] + character_to_add 

#Save to csv to read in the next step, comment out once you do this
# hotel_bookings_messy.to_csv('hotel_bookings_messy.csv')
```

### Addressing NaN

```python
# Replace NaN in 'agent' column with -1 identifier

hotel_bookings['agent'].unique()
hotel_bookings[hotel_bookings['agent'] == 5]

# hotel_bookings['agent'].fillna(-1, inplace=True)
hotel_bookings.fillna({'agent':-1}, inplace=True)
```

```python
# investigating NaN in 'num_children'

hotel_bookings[hotel_bookings['num_children'].isna()] # Only 4 rows, will remove these using the following function

hotel_bookings.dropna(subset = ['num_children'], inplace=True)
```

```python
# investigating NaN in 'country'

hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']] # Check if there are commonalities in the 488 NaN rows; none found

hotel_bookings.fillna({'country':'Unknown'}, inplace=True) # Replacing NaN countries with 'unknown'
```

```python
# drop 'company' column with 94% NaN

hotel_bookings.drop(columns=['company'], inplace=True)
```

### Separate columns
Ex: separate dates

```python
# Split column name by '-' delimiter and select each index to assign to proper new column
hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date'].str.split('-', expand=True)[0]
hotel_bookings['arrival_date_year'] = hotel_bookings['arrival_date'].str.split('-', expand=True)[1]
hotel_bookings.head()    

# Move new columns next to original column
column_to_move = hotel_bookings.pop('arrival_date_month')
hotel_bookings.insert(4, 'arrival_date_month', column_to_move)
hotel_bookings.head() 

column_to_move = hotel_bookings.pop('arrival_date_year')
hotel_bookings.insert(5, 'arrival_date_year', column_to_move)
hotel_bookings.head() 

# Remove original column to avoid duplicate data
hotel_bookings.drop('arrival_date', axis=1, inplace=True)

hotel_bookings.head() 
```
