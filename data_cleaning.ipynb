### Addressing NaN

```python
# Replace NaN in 'agent' column with -1 identifier

hotel_bookings['agent'].unique()
hotel_bookings[hotel_bookings['agent'] == 5]

# hotel_bookings['agent'].fillna(-1, inplace=True)
hotel_bookings.fillna({'agent':-1}, inplace=True)
```

```python
# investigating NaN in 'num_children'

hotel_bookings[hotel_bookings['num_children'].isna()] # Only 4 rows, will remove these using the following function

hotel_bookings.dropna(subset = ['num_children'], inplace=True)
```

```python
# investigating NaN in 'country'

hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']] # Check if there are commonalities in the 488 NaN rows; none found

hotel_bookings.fillna({'country':'Unknown'}, inplace=True) # Replacing NaN countries with 'unknown'
```

```python
# drop 'company' column with 94% NaN

hotel_bookings.drop(columns=['company'], inplace=True)
```

### Check column data types
Can't change NaN values' data type

```python
hotel_bookings.dtypes

# is_canceled is int, should be bool
# num_children stored as float, should be int
# is_repeated_guest is int, should be bool
# agent stored as float, shoule be int
```

```python
hotel_bookings = hotel_bookings.astype({
    'is_canceled': 'boolean',
    'is_repeated_guest': 'boolean',
    'num_children' : 'int',
    'agent' : 'int'})
hotel_bookings.dtypes
```

### Bin columns
Grouping columns by something like a range

```python
hotel_bookings['lead_time'].unique()

hotel_bookings['lead_time'].describe()
```

```python
# Define the bins
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800']

hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels)

hotel_bookings[['lead_time', 'lead_time_binned']]
```
